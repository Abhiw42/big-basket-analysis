### **Step** 8: Find out the outliers from the dataset according to
the columns
and fill them with the mean






1.   FINDING OUTLIERS



```python
cols = ['sale_price', 'market_price', 'rating']

for col in cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```



2.   FIXING OUTLIERS



```python
for col in cols:
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    mean = data[col].mean()

    # Replace outliers with mean
    data[col] = np.where((data[col] < lower_bound) | (data[col] > upper_bound), mean, data[col])
```

```python

print("Outliers identified via boxplots and replaced with column mean.")
```
